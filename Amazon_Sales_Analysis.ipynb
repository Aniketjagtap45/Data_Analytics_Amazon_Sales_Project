```python
data1={"Emp ID":["E01","E02","E03","E04","E05","E06"],
      "Names":["Aniket","Ajay","Avishkar","Rushikesh","Apurv","Shubham"],
      "Age":[24,21,20,24,18,18]}
data2={"Emp ID":["E01","E07","E03","E04","E08","E06"],
      "Emp Salary":[36500,75000,80000,45000,65000,90000]}

df1=pd.DataFrame(data1)
df2=pd.DataFrame(data2)
```
